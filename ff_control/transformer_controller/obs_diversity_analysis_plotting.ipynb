```python
for n_case in range(len(an)):
    print('\n=====================\nCase study : '+ case_study[n_case])
    results = res[case_study[n_case]]    
    results['mean_costs_list'] = []
    results['mean_costs_percgain_list'] = []
    results['mean_iter_list'] = []
    results['mean_iter_percgain_list'] = []
    results['infeas_list'] = []
    results['ctgs0_set'] = []
    results['ctgs0_count'] = []
    scale = 1000
    for i in range(len(an[n_case])):
        print('------------------\nDAGGER --- planning horizon = ', 10*(i+4))
        # Selection for comparison
        J_cvx_comp = results['J_cvx'][i][list(results['ix_ctgs0_ge_1_comparison'][i])].copy()
        J_scp_cvx_comp = results['J_scp_cvx'][i][list(results['ix_ctgs0_ge_1_comparison'][i])].copy()
        J_scp_DT_comp = results['J_scp_DT'][i][list(results['ix_ctgs0_ge_1_comparison'][i])].copy()
        J_scp_DT2_comp = results['J_scp_DT2'][i][list(results['ix_ctgs0_ge_1_comparison'][i])].copy()
        J_scp_DT3_comp = results['J_scp_DT3'][i][list(results['ix_ctgs0_ge_1_comparison'][i])].copy()
        J_scp_DT4_comp = results['J_scp_DT4'][i][list(results['ix_ctgs0_ge_1_comparison'][i])].copy()
        iter_scp_cvx_comp = results['iter_scp_cvx'][i][list(results['ix_ctgs0_ge_1_comparison'][i])].copy()
        iter_scp_DT_comp = results['iter_scp_DT'][i][list(results['ix_ctgs0_ge_1_comparison'][i])].copy()
        iter_scp_DT2_comp = results['iter_scp_DT2'][i][list(results['ix_ctgs0_ge_1_comparison'][i])].copy()
        iter_scp_DT3_comp = results['iter_scp_DT3'][i][list(results['ix_ctgs0_ge_1_comparison'][i])].copy()
        iter_scp_DT4_comp = results['iter_scp_DT4'][i][list(results['ix_ctgs0_ge_1_comparison'][i])].copy()
        ctgs0_cvx_comp = results['ctgs0_cvx'][i][list(results['ix_ctgs0_ge_1_comparison'][i])].copy()

        # Ordered results for ctgs0
        #results['ctgs0_set'].append(set(ctgs0_cvx_comp))
        ctgs0_max = ctgs0_cvx_comp.max()
        results['ctgs0_set'].append(np.arange(0, 51, 1))
        results['mean_costs_list'].append({
            'cvx' : np.zeros((len(results['ctgs0_set'][i],))),
            'scp_cvx' : np.zeros((len(results['ctgs0_set'][i],))),
            'scp_DT' : np.zeros((len(results['ctgs0_set'][i],))),
            'scp_DT2' : np.zeros((len(results['ctgs0_set'][i],))),
            'scp_DT3' : np.zeros((len(results['ctgs0_set'][i],))),
            'scp_DT4' : np.zeros((len(results['ctgs0_set'][i],)))
        })
        results['mean_costs_percgain_list'].append({
            'cvx' : np.zeros((len(results['ctgs0_set'][i],))),
            'scp_cvx' : np.zeros((len(results['ctgs0_set'][i],))),
            'scp_DT' : np.zeros((len(results['ctgs0_set'][i],))),
            'scp_DT2' : np.zeros((len(results['ctgs0_set'][i],))),
            'scp_DT3' : np.zeros((len(results['ctgs0_set'][i],))),
            'scp_DT4' : np.zeros((len(results['ctgs0_set'][i],))),
            'scp_DT4' : np.zeros((len(results['ctgs0_set'][i],)))
        })
        results['mean_iter_list'].append({
            'cvx' :  np.zeros((len(results['ctgs0_set'][i]))),
            'scp_cvx' : np.zeros((len(results['ctgs0_set'][i]))),
            'scp_DT' : np.zeros((len(results['ctgs0_set'][i]))),
            'scp_DT2' : np.zeros((len(results['ctgs0_set'][i]))),
            'scp_DT3' : np.zeros((len(results['ctgs0_set'][i]))),
            'scp_DT4' : np.zeros((len(results['ctgs0_set'][i],)))
        })
        results['mean_iter_percgain_list'].append({
            'cvx' :  np.zeros((len(results['ctgs0_set'][i]))),
            'scp_cvx' : np.zeros((len(results['ctgs0_set'][i]))),
            'scp_DT' : np.zeros((len(results['ctgs0_set'][i]))),
            'scp_DT2' : np.zeros((len(results['ctgs0_set'][i]))),
            'scp_DT3' : np.zeros((len(results['ctgs0_set'][i]))),
            'scp_DT4' : np.zeros((len(results['ctgs0_set'][i],)))
        })
        results['infeas_list'].append({
            'cvx' :  np.zeros((len(results['ctgs0_set'][i]))),
            'scp_cvx' : np.zeros((len(results['ctgs0_set'][i]))),
            'scp_DT' : np.zeros((len(results['ctgs0_set'][i]))),
            'scp_DT2' : np.zeros((len(results['ctgs0_set'][i]))),
            'scp_DT3' : np.zeros((len(results['ctgs0_set'][i]))),
            'scp_DT4' : np.zeros((len(results['ctgs0_set'][i],)))
        })
        results['ctgs0_count'].append(np.zeros((len(results['ctgs0_set'][i],))))

        for j,ctg in enumerate(results['ctgs0_set'][i]):
            idx_ctg = ctgs0_cvx_comp >= ctg
            #idx_ctg = (ctgs0_cvx_comp >= np.floor(ctg/10)*10) & (ctgs0_cvx_comp <= (np.floor(ctg/10) + 1)*10)
            results['mean_costs_list'][i]['cvx'][j] = J_cvx_comp[idx_ctg].mean()

            results['mean_costs_list'][i]['scp_cvx'][j] = J_scp_cvx_comp[idx_ctg].mean()
            results['mean_costs_list'][i]['scp_DT'][j] = J_scp_DT_comp[idx_ctg].mean()
            results['mean_costs_list'][i]['scp_DT2'][j] = J_scp_DT2_comp[idx_ctg].mean()
            results['mean_costs_list'][i]['scp_DT3'][j] = J_scp_DT3_comp[idx_ctg].mean()
            results['mean_costs_list'][i]['scp_DT4'][j] = J_scp_DT4_comp[idx_ctg].mean()
            results['mean_costs_percgain_list'][i]['scp_DT'][j] = (J_scp_cvx_comp[idx_ctg].mean() - J_scp_DT_comp[idx_ctg].mean()) / (J_scp_cvx_comp[idx_ctg].mean() - J_cvx_comp[idx_ctg].mean())
            results['mean_costs_percgain_list'][i]['scp_DT2'][j] = (J_scp_cvx_comp[idx_ctg].mean() - J_scp_DT2_comp[idx_ctg].mean()) / (J_scp_cvx_comp[idx_ctg].mean() - J_cvx_comp[idx_ctg].mean())
            results['mean_costs_percgain_list'][i]['scp_DT3'][j] = (J_scp_cvx_comp[idx_ctg].mean() - J_scp_DT3_comp[idx_ctg].mean()) / (J_scp_cvx_comp[idx_ctg].mean() - J_cvx_comp[idx_ctg].mean())
            results['mean_costs_percgain_list'][i]['scp_DT4'][j] = (J_scp_cvx_comp[idx_ctg].mean() - J_scp_DT4_comp[idx_ctg].mean()) / (J_scp_cvx_comp[idx_ctg].mean() - J_cvx_comp[idx_ctg].mean())

            # Iter
            results['mean_iter_list'][i]['scp_cvx'][j] = iter_scp_cvx_comp[idx_ctg].mean()
            results['mean_iter_list'][i]['scp_DT'][j] = iter_scp_DT_comp[idx_ctg].mean()
            results['mean_iter_list'][i]['scp_DT2'][j] = iter_scp_DT2_comp[idx_ctg].mean()
            results['mean_iter_list'][i]['scp_DT3'][j] = iter_scp_DT3_comp[idx_ctg].mean()
            results['mean_iter_list'][i]['scp_DT4'][j] = iter_scp_DT4_comp[idx_ctg].mean()
            results['mean_iter_percgain_list'][i]['scp_DT'][j] = (iter_scp_cvx_comp[idx_ctg].mean() - iter_scp_DT_comp[idx_ctg].mean()) / (iter_scp_cvx_comp[idx_ctg].mean())
            results['mean_iter_percgain_list'][i]['scp_DT2'][j] = (iter_scp_cvx_comp[idx_ctg].mean() - iter_scp_DT2_comp[idx_ctg].mean()) / (iter_scp_cvx_comp[idx_ctg].mean())
            results['mean_iter_percgain_list'][i]['scp_DT3'][j] = (iter_scp_cvx_comp[idx_ctg].mean() - iter_scp_DT3_comp[idx_ctg].mean()) / (iter_scp_cvx_comp[idx_ctg].mean())
            results['mean_iter_percgain_list'][i]['scp_DT4'][j] = (iter_scp_cvx_comp[idx_ctg].mean() - iter_scp_DT4_comp[idx_ctg].mean()) / (iter_scp_cvx_comp[idx_ctg].mean())
            results['ctgs0_count'][i][j] = idx_ctg.sum()

            # Infeas
            ix_scen = results['scenario_index'][i] == n_case+1
            n_tot_cases_with_ctgs_and_scen_of_interest = len(results['ix'][i][(results['ctgs0_cvx'][i] >= ctg) & ix_scen])
            if n_tot_cases_with_ctgs_and_scen_of_interest > 0:
                results['infeas_list'][i]['scp_cvx'][j] = len(set(results['ix'][i][(results['ctgs0_cvx'][i] >= ctg) & ix_scen]).intersection(set(results['i_unfeas_scp_cvx'][i])))/n_tot_cases_with_ctgs_and_scen_of_interest
                results['infeas_list'][i]['scp_DT'][j] = len(set(results['ix'][i][(results['ctgs0_cvx'][i] >= ctg) & ix_scen]).intersection(set(results['i_unfeas_scp_DT'][i])))/n_tot_cases_with_ctgs_and_scen_of_interest
                results['infeas_list'][i]['scp_DT2'][j] = len(set(results['ix'][i][(results['ctgs0_cvx'][i] >= ctg) & ix_scen]).intersection(set(results['i_unfeas_scp_DT2'][i])))/n_tot_cases_with_ctgs_and_scen_of_interest
                results['infeas_list'][i]['scp_DT3'][j] = len(set(results['ix'][i][(results['ctgs0_cvx'][i] >= ctg) & ix_scen]).intersection(set(results['i_unfeas_scp_DT3'][i])))/n_tot_cases_with_ctgs_and_scen_of_interest
                results['infeas_list'][i]['scp_DT4'][j] = len(set(results['ix'][i][(results['ctgs0_cvx'][i] >= ctg) & ix_scen]).intersection(set(results['i_unfeas_scp_DT4'][i])))/n_tot_cases_with_ctgs_and_scen_of_interest
            else:
                results['infeas_list'][i]['scp_cvx'][j] = 0
                results['infeas_list'][i]['scp_DT'][j] = 0
                results['infeas_list'][i]['scp_DT2'][j] = 0
                results['infeas_list'][i]['scp_DT3'][j] = 0
                results['infeas_list'][i]['scp_DT4'][j] = 0
```
