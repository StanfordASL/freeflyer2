# FINAL TIME Analysis

```python
an = [[]]
case_study = ['freeflyer']
name_file = [root_folder + '/optimization/saved_files/warmstarting/ws_analysisFULL_checkpoint_ff_time40_100_chunk100R_ctgrtg12000_comp_dyn']
ttg_i = 40
for i in range(len(an)):
    an[i].append(np.load(name_file[i] + '40.0.npz'))
    an[i].append(np.load(name_file[i] + '50.0.npz'))
    an[i].append(np.load(name_file[i] + '60.0.npz'))
    an[i].append(np.load(name_file[i] + '70.0.npz'))
    an[i].append(np.load(name_file[i] + '80.0.npz'))
    an[i].append(np.load(name_file[i] + '90.0.npz'))
    an[i].append(np.load(name_file[i] + '100.0.npz'))
```

```python
res_dict = {
    'ix_sample' : np.array([]),
    'ix' : np.array([]),
    'J_cvx' : np.array([]),
    'J_scp_cvx' : np.array([]),
    'J_scp_DT' : np.array([]),
    'J_scp_DT2' : np.array([]),
    'J_scp_DT3' : np.array([]),
    'iter_scp_cvx' : np.array([]),
    'iter_scp_DT' : np.array([]),
    'iter_scp_DT2' : np.array([]),
    'iter_scp_DT3' : np.array([]),
    'ctgs0_cvx' : np.array([]),
    'cvx_problem' : np.array([]),
    'i_unfeas_cvx' : np.array([]),
    'i_unfeas_scp_cvx' : np.array([]),
    'i_unfeas_scp_DT' : np.array([]),
    'i_unfeas_scp_DT2' : np.array([]),
    'i_unfeas_scp_DT3' : np.array([]),
    'final_time' : np.array([])
}
res = {
    'freeflyer' : copy.deepcopy(res_dict)
}

for n_case in range(len(an)):
    results = res[case_study[n_case]]
    for i in range(len(an[n_case])):
        results['ix_sample'] = np.append(results['ix_sample'], an[n_case][i]['test_dataset_ix'])
        results['ix'] = np.append(results['ix'].astype(int), np.arange(len(an[n_case][i]['test_dataset_ix'])).astype(int) + int(5000*i))
        results['J_cvx'] = np.append(results['J_cvx'], an[n_case][i]['J_cvx'])
        results['J_scp_cvx'] = np.append(results['J_scp_cvx'], np.min(an[n_case][i]['J_vect_scp_cvx'], axis=1))
        results['J_scp_DT'] = np.append(results['J_scp_DT'], np.min(an[n_case][i]['J_vect_scp_DT'], axis=1))
        results['J_scp_DT2'] = np.append(results['J_scp_DT2'], np.min(an[n_case][i]['J_vect_scp_DT2'], axis=1))
        results['J_scp_DT3'] = np.append(results['J_scp_DT3'], np.min(an[n_case][i]['J_vect_scp_DT3'], axis=1))
        results['iter_scp_cvx'] = np.append(results['iter_scp_cvx'], an[n_case][i]['iter_scp_cvx'])
        results['iter_scp_DT'] = np.append(results['iter_scp_DT'], an[n_case][i]['iter_scp_DT'])
        results['iter_scp_DT2'] = np.append(results['iter_scp_DT2'], an[n_case][i]['iter_scp_DT2'])
        results['iter_scp_DT3'] = np.append(results['iter_scp_DT3'], an[n_case][i]['iter_scp_DT3'])
        results['cvx_problem'] = np.append(results['cvx_problem'], an[n_case][i]['cvx_problem'])
        results['i_unfeas_cvx'] = np.append(results['i_unfeas_cvx'].astype(int), an[n_case][i]['i_unfeas_cvx'].astype(int) + int(5000*i))
        results['i_unfeas_scp_cvx'] = np.append(results['i_unfeas_scp_cvx'].astype(int), an[n_case][i]['i_unfeas_scp_cvx'].astype(int) + int(5000*i))
        results['i_unfeas_scp_DT'] = np.append(results['i_unfeas_scp_DT'].astype(int), an[n_case][i]['i_unfeas_scp_DT'].astype(int) + int(5000*i))
        results['i_unfeas_scp_DT2'] = np.append(results['i_unfeas_scp_DT2'].astype(int), an[n_case][i]['i_unfeas_scp_DT2'].astype(int) + int(5000*i))
        results['i_unfeas_scp_DT3'] = np.append(results['i_unfeas_scp_DT3'].astype(int), an[n_case][i]['i_unfeas_scp_DT3'].astype(int) + int(5000*i))
        results['final_time'] = np.append(results['final_time'], (i*10 + ttg_i)*np.ones_like(an[n_case][i]['test_dataset_ix']))
        results['ctgs0_cvx'] = np.append(results['ctgs0_cvx'], an[n_case][i]['ctgs0_cvx'] / ((i*10 + ttg_i)*np.ones_like(an[n_case][i]['test_dataset_ix'])))
```

```python
for n_case in range(len(an)):
    print('\n=====================\nCase study : '+ case_study[n_case])
    results = res[case_study[n_case]]
    results['ix_ctgs0_ge_1_comparison'] = []
    print('------------------\nDAGGER --- final time = MIXED')
    ix_unfeas = set(results['i_unfeas_cvx']).union(set(results['i_unfeas_scp_cvx'])).union(set(results['i_unfeas_scp_DT'])).union(set(results['i_unfeas_scp_DT2']))
    #ix_unfeas = ix_unfeas.union(set(results['i_unfeas_scp_DT3']))
    ix_unfeas3 = ix_unfeas
    ix_feas_comparison = set(results['ix']) - ix_unfeas - set([-1])
    ix_feas_comparison3 = set(results['ix']) - ix_unfeas3 - set([-1])
    print('Feasible samples for comparison', len(ix_feas_comparison))
    ctgs_selector = (results['ctgs0_cvx'] >= 0) & (results['ctgs0_cvx'] <= 1)
    ix_ctgs0_ge_1 = set(results['ix'][ctgs_selector])
    print('Samples with ctgs0 of interest 0<=ctgs0<='+str(10*(i+4)), len(ix_ctgs0_ge_1))
    results['ix_ctgs0_ge_1_comparison'] = ix_feas_comparison.intersection(ix_ctgs0_ge_1)#.intersection(set(results['ix'][results['final_time'] == 50.0]))
    results['ix_ctgs0_ge_1_comparison3'] = ix_feas_comparison3.intersection(ix_ctgs0_ge_1)#.intersection(set(results['ix'][results['final_time'] == 50.0]))
    print('Samples with ctgs0 of insterest for comparison', len(results['ix_ctgs0_ge_1_comparison']))
```

```python
for n_case in range(len(an)):
    print('\n=====================\nCase study : '+ case_study[n_case])
    results = res[case_study[n_case]]    
    results['mean_costs_list'] = []
    results['mean_costs_percgain_list'] = []
    results['mean_iter_list'] = []
    results['infeas_list'] = []
    results['ctgs0_set'] = []
    results['ctgs0_set3'] = []
    results['ctgs0_count'] = []
    scale = 1000
    print('------------------\nDAGGER --- final time = MIXED')
    # Selection for comparison
    J_cvx_comp = results['J_cvx'][list(results['ix_ctgs0_ge_1_comparison'])].copy()
    J_cvx_comp3 = results['J_cvx'][list(results['ix_ctgs0_ge_1_comparison3'])].copy()
    J_scp_cvx_comp = results['J_scp_cvx'][list(results['ix_ctgs0_ge_1_comparison'])].copy()
    J_scp_cvx_comp3 = results['J_scp_cvx'][list(results['ix_ctgs0_ge_1_comparison3'])].copy()
    J_scp_DT_comp = results['J_scp_DT'][list(results['ix_ctgs0_ge_1_comparison'])].copy()
    J_scp_DT2_comp = results['J_scp_DT2'][list(results['ix_ctgs0_ge_1_comparison'])].copy()
    J_scp_DT3_comp = results['J_scp_DT3'][list(results['ix_ctgs0_ge_1_comparison3'])].copy()
    iter_scp_cvx_comp = results['iter_scp_cvx'][list(results['ix_ctgs0_ge_1_comparison'])].copy()
    iter_scp_DT_comp = results['iter_scp_DT'][list(results['ix_ctgs0_ge_1_comparison'])].copy()
    iter_scp_DT2_comp = results['iter_scp_DT2'][list(results['ix_ctgs0_ge_1_comparison'])].copy()
    iter_scp_DT3_comp = results['iter_scp_DT3'][list(results['ix_ctgs0_ge_1_comparison3'])].copy()
    ctgs0_cvx_comp = results['ctgs0_cvx'][list(results['ix_ctgs0_ge_1_comparison'])].copy()
    ctgs0_cvx_comp3 = results['ctgs0_cvx'][list(results['ix_ctgs0_ge_1_comparison3'])].copy()

    # Ordered results for ctgs0
    results['ctgs0_set'].append(np.unique(ctgs0_cvx_comp))
    results['ctgs0_set3'].append(np.unique(ctgs0_cvx_comp3))
    results['mean_costs_list'].append({
        'cvx' : np.zeros((len(results['ctgs0_set'][0],))),
        'cvx3' : np.zeros((len(results['ctgs0_set3'][0],))),
        'scp_cvx' : np.zeros((len(results['ctgs0_set'][0],))),
        'scp_DT' : np.zeros((len(results['ctgs0_set'][0],))),
        'scp_DT2' : np.zeros((len(results['ctgs0_set'][0],))),
        'scp_DT3' : np.zeros((len(results['ctgs0_set3'][0],)))
    })
    results['mean_costs_percgain_list'].append({
        'scp_DT' : np.zeros((len(results['ctgs0_set'][0],))),
        'scp_DT2' : np.zeros((len(results['ctgs0_set'][0],))),
        'scp_DT3' : np.zeros((len(results['ctgs0_set3'][0],)))
    })
    results['mean_iter_list'].append({
        'cvx' :  np.zeros((len(results['ctgs0_set'][0]))),
        'cvx3' :  np.zeros((len(results['ctgs0_set3'][0]))),
        'scp_cvx' : np.zeros((len(results['ctgs0_set'][0]))),
        'scp_DT' : np.zeros((len(results['ctgs0_set'][0]))),
        'scp_DT2' : np.zeros((len(results['ctgs0_set'][0]))),
        'scp_DT3' : np.zeros((len(results['ctgs0_set3'][0])))
    })
    results['infeas_list'].append({
        'cvx' :  np.zeros((len(results['ctgs0_set'][0]))),
        'cvx3' :  np.zeros((len(results['ctgs0_set3'][0]))),
        'scp_cvx' : np.zeros((len(results['ctgs0_set'][0]))),
        'scp_DT' : np.zeros((len(results['ctgs0_set'][0]))),
        'scp_DT2' : np.zeros((len(results['ctgs0_set'][0]))),
        'scp_DT3' : np.zeros((len(results['ctgs0_set3'][0])))
    })
    results['ctgs0_count'].append(np.zeros((len(results['ctgs0_set'][0],))))

    for j,ctg in enumerate(results['ctgs0_set'][0]):
        idx_ctg = ctgs0_cvx_comp >= ctg#(ctgs0_cvx_comp >= ctg - 0.1) & (ctgs0_cvx_comp <= ctg + 0.1)
        #idx_ctg = (ctgs0_cvx_comp >= np.floor(ctg/10)*10) & (ctgs0_cvx_comp <= (np.floor(ctg/10) + 1)*10)
        results['mean_costs_list'][0]['cvx'][j] = J_cvx_comp[idx_ctg].mean()

        results['mean_costs_list'][0]['scp_cvx'][j] = J_scp_cvx_comp[idx_ctg].mean()
        results['mean_costs_list'][0]['scp_DT'][j] = J_scp_DT_comp[idx_ctg].mean()
        results['mean_costs_list'][0]['scp_DT2'][j] = J_scp_DT2_comp[idx_ctg].mean()
        #results['mean_costs_list'][0]['scp_DT3'][j] = J_scp_DT3_comp[idx_ctg].mean()

        results['mean_costs_percgain_list'][0]['scp_DT'][j] = (results['mean_costs_list'][0]['scp_cvx'][j] - results['mean_costs_list'][0]['scp_DT'][j]) / (
                                                                results['mean_costs_list'][0]['scp_cvx'][j] - results['mean_costs_list'][0]['cvx'][j])
        results['mean_costs_percgain_list'][0]['scp_DT2'][j] = (results['mean_costs_list'][0]['scp_cvx'][j] - results['mean_costs_list'][0]['scp_DT2'][j]) / (
                                                                results['mean_costs_list'][0]['scp_cvx'][j] - results['mean_costs_list'][0]['cvx'][j])
        #results['mean_costs_percgain_list'][0]['scp_DT3'][j] = (results['mean_costs_list'][0]['scp_cvx'][j] - results['mean_costs_list'][0]['scp_DT3'][j]) / (
        #                                                        results['mean_costs_list'][0]['scp_cvx'][j] - results['mean_costs_list'][0]['cvx'][j])

        # Iter
        results['mean_iter_list'][0]['scp_cvx'][j] = iter_scp_cvx_comp[idx_ctg].mean()
        results['mean_iter_list'][0]['scp_DT'][j] = iter_scp_DT_comp[idx_ctg].mean()
        results['mean_iter_list'][0]['scp_DT2'][j] = iter_scp_DT2_comp[idx_ctg].mean()
        #results['mean_iter_list'][0]['scp_DT3'][j] = iter_scp_DT3_comp[idx_ctg].mean()
        results['ctgs0_count'][0][j] = idx_ctg.sum()

        # Infeas
        idx_inf_ctg_ft = set(results['ix'][results['ctgs0_cvx'] >= ctg])#.intersection(set(results['ix'][results['final_time'] == 40.0]))
        results['infeas_list'][0]['scp_cvx'][j] = len(idx_inf_ctg_ft.intersection(set(results['i_unfeas_scp_cvx'])))/len(idx_inf_ctg_ft)
        results['infeas_list'][0]['scp_DT'][j] = len(idx_inf_ctg_ft.intersection(set(results['i_unfeas_scp_DT'])))/len(idx_inf_ctg_ft)
        results['infeas_list'][0]['scp_DT2'][j] = len(idx_inf_ctg_ft.intersection(set(results['i_unfeas_scp_DT2'])))/len(idx_inf_ctg_ft)
        #results['infeas_list'][0]['scp_DT3'][j] = len(idx_inf_ctg_ft.intersection(set(results['i_unfeas_scp_DT3'])))/len(idx_inf_ctg_ft)
    
    for j,ctg in enumerate(results['ctgs0_set3'][0]):
        idx_ctg = ctgs0_cvx_comp3 >= ctg#(ctgs0_cvx_comp3 >= ctg - 0.1) & (ctgs0_cvx_comp3 >= ctg + 0.1)
        #idx_ctg = (ctgs0_cvx_comp >= np.floor(ctg/10)*10) & (ctgs0_cvx_comp <= (np.floor(ctg/10) + 1)*10)
        results['mean_costs_list'][0]['cvx3'][j] = J_cvx_comp3[idx_ctg].mean()

        results['mean_costs_list'][0]['scp_DT3'][j] = J_scp_DT3_comp[idx_ctg].mean()
        results['mean_costs_percgain_list'][0]['scp_DT3'][j] = (J_scp_cvx_comp3[idx_ctg].mean() - J_scp_DT3_comp[idx_ctg].mean()) / (
                                                                J_scp_cvx_comp3[idx_ctg].mean() - J_cvx_comp3[idx_ctg].mean())

        # Iter
        results['mean_iter_list'][0]['scp_DT3'][j] = iter_scp_DT3_comp[idx_ctg].mean()

        # Infeas
        idx_inf_ctg_ft = set(results['ix'][results['ctgs0_cvx'] >= ctg])#.intersection(set(results['ix'][results['final_time'] == 40.0]))
        results['infeas_list'][0]['scp_DT3'][j] = len(idx_inf_ctg_ft.intersection(set(results['i_unfeas_scp_DT3'])))/len(idx_inf_ctg_ft)
```
